linear equation
y = mx + c, where 
m => slope
c => intercept

Getting all libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
plt.rcParams['figure.figsize'] = (12.0,9.0)
```

Loading Data Frame

```python
df = pd.read_csv("HousePrice.csv")
X = df.iloc[:, 1]
Y = df.iloc[:, 0]
```

```python
plt.xlabel('Price')
plt.ylabel('Area')
plt.scatter(X,Y)
plt.show()
```

```python
#Building Model
m = 0 
c = 0
L = 0.0001
n = float(len(X))
for i in range(len(X)):
    Y_Pred = m * X + c
    #cost function
    cost = sum([data**2 for data in (Y-Y_Pred)]) / n
    D_m = (-2/n) * sum(X * (Y - Y_Pred))
    D_c = (-2/n) * sum(Y - Y_Pred)
    m = m - L*D_m
    c = c - L*D_c
    print(cost)
print(cost)
```

```python
Y_Pred = m*X + c
plt.xlabel('Price')
plt.ylabel('Area')
plt.scatter(X,Y)
plt.plot([min(X), max(X)], [min(Y_Pred),max(Y_Pred)], color='red')
plt.show()
```

```python
#Price of 1000 sqrt appartment is 
Xp = (1000 - c) / m
print(round(Xp,2),"Lakhs")

#Area for 100 lakh appartment is
Yp = 100*m + c
print(round(Yp,2)," sqr Foot")
```
